```python
## Bike Trip Data Analysis Notebook

# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set styles
sns.set(style="whitegrid")

# 2. Load Dataset
df = pd.read_csv('201801-fordgobike-tripdata.csv')
df.head()

# ---
# 1. Overview of Bike Trips
print("\nDataset Info:\n")
df.info()

print("\nSummary Statistics:\n")
print(df.describe())

print("\nMissing Values:\n")
print(df.isnull().sum())

# ---
# 2. Station Analysis

# Top start stations
top_start_stations = df['start_station_name'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_start_stations.values, y=top_start_stations.index, palette="viridis")
plt.title('Top 10 Start Stations')
plt.xlabel('Number of Trips')
plt.ylabel('Start Station')
plt.show()

# Top end stations
top_end_stations = df['end_station_name'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_end_stations.values, y=top_end_stations.index, palette="magma")
plt.title('Top 10 End Stations')
plt.xlabel('Number of Trips')
plt.ylabel('End Station')
plt.show()

# ---
# 3. Trip Duration Analysis

# Trip duration distribution
plt.figure(figsize=(10,6))
sns.histplot(df['duration_sec']/60, bins=50, kde=True, color='skyblue')
plt.xlim(0,60)
plt.title('Trip Duration Distribution (Minutes)')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Frequency')
plt.show()

# Average trip duration by user type
plt.figure(figsize=(8,5))
sns.barplot(x='user_type', y='duration_sec', data=df, estimator=lambda x: sum(x) / len(x), palette="pastel")
plt.title('Average Trip Duration by User Type')
plt.ylabel('Average Duration (Seconds)')
plt.show()

# ---
# 4. Hot Analysis (Popular Times & Stations)

# Convert to datetime
df['start_time'] = pd.to_datetime(df['start_time'])

# Extract hour
df['hour'] = df['start_time'].dt.hour

# Rush hours
plt.figure(figsize=(10,6))
sns.countplot(x='hour', data=df, palette="coolwarm")
plt.title('Number of Trips by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Trips')
plt.show()

# Most active days
df['day_of_week'] = df['start_time'].dt.day_name()
plt.figure(figsize=(10,6))
sns.countplot(x='day_of_week', data=df, order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], palette="Spectral")
plt.title('Number of Trips by Day of Week')
plt.xlabel('Day')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.show()
```
